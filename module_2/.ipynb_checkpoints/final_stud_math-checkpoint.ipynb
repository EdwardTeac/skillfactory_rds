```python
# Ищем все индексы где остсутствуют данные
medu_nan = students.loc[pd.isna(students['Medu']), :].index
fedu_nan = students.loc[pd.isna(students['Fedu']), :].index
freetime_nan = students.loc[pd.isna(students['freetime']), :].index
goout_nan = students.loc[pd.isna(students['goout']), :].index


# Обычно пары соответсвуют образованию друг друга, заполняем столбец Medu на основе Fedu
for i in medu_nan:
    students['Medu'][i] = students['Fedu'][i]
```

```python
# чистим выбранные столбцы с выбросами
selected_columns = ['Fedu','traveltime','famrel','studytime','absences','freetime','score']

for col in selected_columns:
    perc25 = students.loc[:, col].quantile(0.25) # Высчитываем Q1
    perc75 = students.loc[:, col].quantile(0.75) # Высчитываем Q1
    IQR =  perc75 -  perc25 # Высчитываем IQR
    students = students.loc[students.loc[:, col].between(perc25 - 1.5*IQR, perc75 + 1.5*IQR)] # Фильтруем выбросы
```

```python
# После чистки перестали отсутсвовать пустые строки в столбцах с числами
# Убираем не нужные столбцы, которые меньше всего влияют на score

students.drop(['famrel'], inplace = True, axis = 1)
students.drop(['freetime'], inplace = True, axis = 1)
students.drop(['health'], inplace = True, axis = 1)
students.drop(['absences'], inplace = True, axis = 1)
students.drop(['traveltime'], inplace = True, axis = 1)

# STUDYTIME,GRANULAR и studytime полностью скорелированы, убираем studytime
students.drop(['studytime'], inplace = True, axis = 1)
```

```python
# Заполняем оставшие нумеративные столбцы наиболее частыми данными
non_num = ['address','famsize','Pstatus','Mjob','Fjob','reason','guardian','schoolsup',
           'famsup','paid','activities','nursery','higher','internet','romantic']

for col in non_num:
    students.loc[:, col].fillna(students.loc[:, col].value_counts().index.tolist()[0],inplace=True)
```

```python
# Прверим не нумеративные данные по отношению к score

students.reset_index()
for column in non_num:
    fig, ax = plt.subplots(figsize = (14, 4))
    sns.boxplot(x=column, y='score', 
                data=students.loc[students.loc[:, column].isin(students.loc[:, column].value_counts().index[:10])],
               ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('Boxplot for ' + column)
    plt.show()
```

```python
# Уберем столбцы которые не влияют на score

students.drop(['famsize'], inplace = True, axis = 1)
students.drop(['Pstatus'], inplace = True, axis = 1)
students.drop(['famsup'], inplace = True, axis = 1)
students.drop(['paid'], inplace = True, axis = 1)
students.drop(['activities'], inplace = True, axis = 1)
students.drop(['nursery'], inplace = True, axis = 1)
```

```python
non_num = ['address','Mjob','Fjob','reason','guardian','schoolsup',
           'higher','internet','romantic']

# Проверим, есть ли статистическая разница в распределении оценок по номинативным признакам
def get_stat_dif(column):
    cols = students.loc[:, column].value_counts().index[:10]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if ttest_ind(students.loc[students.loc[:, column] == comb[0], 'score'], 
                        students.loc[students.loc[:, column] == comb[1], 'score']).pvalue \
            <= 0.05/len(combinations_all):
            print('Найдены статистически значимые различия для колонки', column)
            break
for col in non_num:
    get_stat_dif(col)
```
