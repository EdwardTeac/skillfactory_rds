```python
import pandas as pd
ratings = pd.read_csv('ratings_example.txt', sep = '\t')
ratings.head()
```

```python
movies = pd.read_csv('movies_example.txt', sep = '\t')
movies.head()
```

```python
merged = ratings.merge(movies, on='movieId', how='left')
print('Объединение с лева: \n{} \n'.format(merged))

merged = ratings.merge(movies, on='movieId', how='right')
print('Объединение с права: \n {} \n'.format(merged) )

merged = ratings.merge(movies, on='movieId', how='inner')
print('Объединение с inner: \n {} \n'.format(merged) )

merged = ratings.merge(movies, on='movieId', how='outer')
print('Объединение с outer: \n {} \n'.format(merged) )
```

```python
items_dict = {

    'item_id': [417283, 849734, 132223, 573943, 19475, 3294095, 382043, 302948, 100132, 312394], 

    'vendor': ['Samsung', 'LG', 'Apple', 'Apple', 'LG', 'Apple', 'Samsung', 'Samsung', 'LG', 'ZTE'],

    'stock_count': [54, 33, 122, 18, 102, 43, 77, 143, 60, 19]

}
purchase_log = {

    'purchase_id': [101, 101, 101, 112, 121, 145, 145, 145, 145, 221],

    'item_id': [417283, 849734, 132223, 573943, 19475, 3294095, 382043, 302948, 103845, 100132], 

    'price': [13900, 5330, 38200, 49990, 9890, 33000, 67500, 34500, 89900, 11400]

}
items_df = pd.DataFrame(items_dict)
purchase_df = pd.DataFrame(purchase_log)
```

```python
merged = purchase_df.merge(items_df, on='item_id', how='outer')
merged
```

```python
merged = purchase_df.merge(items_df, on='item_id', how='inner')
merged
```

```python
# Добавили столбец "Profit", умножив price на stock_count
merged['profit'] = merged.apply(lambda row: row.price * row.stock_count, axis=1)
new_merged = merged.sort_values(['profit'], ascending=False)
display(new_merged)

print('Итоговавя выручка по всем моделям -\n \033[1m{}\033[0m\n'
      .format(new_merged.profit.sum()))
```

```python
# Собираем содержимое файлов папки data в единый датафрейм data
files = os.listdir('data')
data = pd.DataFrame(columns = ['userId', 'movieId', 'rating', 'timestamp'])

for filename in files:
    temp = pd.read_csv( os.path.join('data', filename), names = ['userId', 'movieId', 'rating', 'timestamp'] )
    data = pd.concat([data, temp])
    
data.reset_index()
```
