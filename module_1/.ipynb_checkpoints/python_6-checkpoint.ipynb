```python
df = pd.read_csv('data_sf.xls')

spains = df.loc[df['Nationality'] == 'Spain'].Wage.value_counts(bins=4,normalize=True) 
# Собираем отдельно Испанцев и сразу же разделяем их на 4 группы по зарплатам и указываем количество в %
print('Группы по 25% с зарплатами, и количество Испанцев в процентах -\n \033[1m{}\033[0m\n'
      .format(spains))

manchester = df.loc[df['Club'] == 'Manchester United'].Nationality.nunique()
print('Количество уникальных сборных (Nationality) из футболистов Manchester United - \033[1m{}\033[0m\n'
      .format(manchester))

juventus_brazil = df.loc[(df.Club == 'Juventus') & (df.Nationality == 'Brazil')].Name.unique()
print('Футболисты из бразилии играющие за Ювентус - \033[1m{}\033[0m\n'
      .format(juventus_brazil))

oldest_wage_club = df.loc[df.Age > 35].Club.value_counts()
print('Клуб (Club) который, насчитывает большее количество футболистов возрастом (Age) старше 35 лет - \033[1m{}\033[0m\n'
      .format(oldest_wage_club.index.values[0]))

argentina_age_group = df.loc[df.Nationality == 'Argentina'].Age.value_counts(bins=4)
print('Футболисты сборной Аргентины в 4х группах по возрасту -\n \033[1m{}\033[0m\n'
      .format(argentina_age_group))

young_age = df.loc[df.Nationality == 'Spain'].Age.value_counts(normalize=True)
young_age_rounded = young_age.replace(young_age.values, young_age.values*100).round(decimals=2)[0:5]

print('Проценты фубтолистов из Испании по возрастам -\n \033[1m{}\033[0m\n'
      .format(young_age_rounded))
```

```python
df = pd.read_csv('data_sf.xls')

grouped_df = df.groupby(['Club']).sum()
display(grouped_df)
display(grouped_df.loc['Ajax']['Wage'])
grouped_df = df.groupby(['Club'])['Wage'].sum().sort_values(ascending=False)
print(grouped_df.head(5))
grouped_df = df.groupby(['Position'])['Wage'].sum().sort_values(ascending=False)
print(grouped_df.head(10))
```

```python
# Сортируем список по национальностям с средними значениями 'Wage','Age','ShotPower' по Зарплате 'Wage' в порядке убывания
display(df.groupby(['Nationality'])[['Wage','Age','ShotPower']].mean().sort_values(['Wage'],ascending=False).head(10))
# Смотрим кто входит в Dominican Republic с выборочними столбцами
display(df.loc[df['Nationality'] == 'Dominican Republic'][['Name','Club','Wage','Age','ShotPower']])
```

```python
df = pd.read_csv('data_sf.xls')
df.groupby(['Position'])[['Wage','Value']].mean().sort_values(['Value'],ascending=False).head(10).round(2)
```

```python
# Средняя (mean) и медианная (median) зарплат (Wage),которые совпадают, футболистов из разных клубов (Club)
temp = df.groupby(['Club'])['Wage'].agg(['mean', 'median'])
display(temp.loc[temp['mean'] == temp['median']].sort_values(['mean'],ascending=False).reset_index()[0:10])

```

```python
tmp_chelsea = df.groupby(['Club'])['Wage'].sum()
print('Сумма зарплат (Wage) футболистов клуба (Club) "Chelsea" - \033[1m{}\033[0m\n'
      .format(tmp_chelsea['Chelsea']))

max_wage_20 = df.loc[(df['Age'] == 20) & (df['Nationality'] == 'Argentina')]['Wage'].max()
print('Максимальная зарплата футболиста национальности (Nationality) Аргентина ("Argentina") в возрасте 20 лет - \033[1m{}\033[0m\n'
      .format(max_wage_20))

max_wage_30 = df.loc[(df['Age'] == 30) & (df['Nationality'] == 'Argentina')]['Wage'].max()
print('Максимальная зарплата футболиста национальности (Nationality) Аргентина ("Argentina") в возрасте 30 лет - \033[1m{}\033[0m\n'
      .format(max_wage_30))

min_wage_30 = df.loc[(df['Age'] == 30) & (df['Nationality'] == 'Argentina')]['Wage'].min()
print('Максимальная зарплата футболиста национальности (Nationality) Аргентина ("Argentina") в возрасте 30 лет - \033[1m{}\033[0m\n'
      .format(min_wage_30))

str_bal = df.loc[(df['Club'] == 'FC Barcelona') & (df['Nationality'] == 'Argentina')][['Strength','Balance']].max()
print('Максимальная сила (Strength) и баланс (Balance) среди игроков клуба (Club) "FC Barcelona" из Аргентины ("Argentina") -\n\033[1m{}\033[0m\n'
      .format(str_bal))
```

```python
pivot = df.loc[df['Club'].isin(['FC Barcelona','Real Madrid','Juventus','Manchester United'])].pivot_table(values=['Wage'],
index=['Nationality'],
columns=['Club'],
aggfunc='sum', # aggfunc Добавляеи функцию к параметрам
margins=True, # Margings Значение "Всего" по краям
fill_value=0) # Заменяет все NaN в таблице на 0
display(pivot)
```

```python
pivot = df.loc[df['Club'].isin(['FC Barcelona','Real Madrid','Juventus','Manchester United'])].pivot_table(values=['Name'],
index=['Nationality'],
columns=['Club'],
aggfunc='count',
fill_value=0)
display(pivot)
pivot.loc['Argentina']['Name']['Real Madrid']
```

```python
pivot = df.pivot_table(values=['Name'],
index=['Club'],
columns=['Position'],
aggfunc='count',
fill_value = 0)
display(pivot)

print('Среднее количество вратарей (GK) - \033[1m{}\033[0m\n'
      .format(round(pivot.Name.GK.mean(),3)))
```

```python
print('Клубы, которые, не содержат данных о центральных полузащитниках (CM) -\n \033[1m{}\033[0m\n'
      .format(pivot.loc[pivot.Name.CM == 0].Name.CM.value_counts()))
```

```python
pivot = df.loc[df['Club'].isin(['AS Monaco'])].pivot_table(values=['Wage'],
index=['Nationality'],
columns=['Club'],
aggfunc='sum',
fill_value = 0)
print('Сколько получают ("Wage") русские футболисты ("Russia"), играющие за ФК "AS Monaco" -\n \033[1m{}\033[0m\n'
      .format(pivot))
```

```python
pivot = df.pivot_table(values=['SprintSpeed'],
index=['Position'],
columns=['Club'],
aggfunc='mean',
margins=True,
fill_value = 0)
pivot.sort_values(by=('SprintSpeed','All'),ascending=False)
```

```python
pivot = df.pivot_table(values=['SprintSpeed'],
index=['Position'],
columns=['Club'],
aggfunc='mean',
margins=True,
fill_value = 0)
pivot.sort_values(['ST'],axis=1,ascending=False) # Axis = 1 Сортировка по индексам а не по столбцам
```
