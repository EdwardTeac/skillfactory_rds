```python
math = []
read = []
read_girl = []
write_count_students = 0
write_count_students_lunch = 0

f = open('StudentsPerformance.xls')

for line in f:
    info = line.split(',')
    if info[0] == '"gender"':
        continue
    else:
        mark = int(info[5][1:-1])
        mark_read = int(info[-2][1:-1])
        mark_write =int(info[-1][1:-2])
        if mark_write > 90 and info[3][1:-1] == 'standard':
            write_count_students_lunch += 1
        
        if mark_write > 90:
            write_count_students += 1
        read.append(mark_read)
        math.append(mark)
        if info[0][1:-1] == 'female':
            read_girl.append(mark_read)
            
f.close()
math_avg = sum(math)/len(math)
read_avg = sum(read)/len(read)
read_avg_girls = round(sum(read_girl)/len(read_girl),3)


print('Общие средние балы: Математика: {}, Чтение: {}'.format(math_avg,read_avg))
print('Количество студентов с оценкой выше 90 по Письму: {}'.format(write_count_students))
print('Процент студентов с оценкой выше 90 по Письму,которые хорошо пообедали: {}'.format(round((write_count_students_lunch*100)/write_count_students,1)))
above_avg_math = 0
under_avg_read = 0
for mark in math:
    if mark > math_avg:
        above_avg_math += 1
for mark in read:
    if mark < read_avg:
        under_avg_read += 1
print('Показатели выше среднего- Математика: {}, Показатели ниже среднего Чтение: {}'.format(above_avg_math / len(math),under_avg_read / len(read)))
print('Средний показатель по чтению у девочек: - {}'.format(read_avg_girls))
```

```python
import re
import pprint
pattern = re.compile('\d+')

exams = []

f = open('StudentsPerformance.xls')

for line in f:
    info = line.split(',')
    if info[0] == '"gender"':
        continue
    else:
        new_line = []
        for item in info:
            if pattern.search(item) != None:
                new_line.append(int(pattern.search(item)[0]))
            else:
                new_line.append(item[1:-1])
        exams.append(new_line)
        
pp = pprint.PrettyPrinter(width=120,sort_dicts=False)
pp.pprint(exams)
```

```python
import collections
import pprint
students = {}

f = open('StudentsPerformance.xls')
c = collections.Counter()
b = collections.Counter()
d = collections.Counter()
e = collections.Counter()
count = 0
count_group_c = 0
for line in f:
    info = line.split(',')
    if info[0] == '"gender"':
        continue
    else: 
        if info[0][1:-1] == 'male':
            students['Мальчики'] = {}
            food = info[3][1:-1]
            tests = info[4][1:-1]
            if food == 'standard':
                c[food] +=1
            if tests == 'completed':
                b[tests] +=1
        elif info[0][1:-1] == 'female':
            students['Девочки'] = {}
            parent_study = info[2][1:-1]
            if parent_study == "master's degree":
                d[parent_study] +=1
            if parent_study == "master's degree" and int(info[-3][1:-1]) > 90:
                count +=1
        if info[1][1:-1] == 'group C' and info[4][1:-1] == 'completed':
            count_group_c += 1
            
else:   
    f.close()
    students['Все'] = {}
    students['Мальчики']['Обед передет экзаменом'] = dict(c)
    students['Мальчики']['Подготовительные курсы'] = dict(b)
    students['Девочки']['Образование родителей'] = dict(d)
    students['Девочки']['Больше 90 баллов'] = count
    students['Все']['Группа C с подготовиьельными курсами'] = count_group_c
    pp.pprint(students)
```
